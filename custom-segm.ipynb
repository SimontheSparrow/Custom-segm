```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://storage.googleapis.com/esmartdata-courses-files/ml-course/OnlineRetail.csv', encoding='latin', parse_dates=['InvoiceDate'])
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
tmp = df['Country'].value_counts().reset_index()
tmp.columns = ['Country', 'Count']
tmp.query("Count > 200", inplace=True)


df_uk = df.query("Country == 'United Kingdom'").copy()
df_uk.head()
```

```python
df_uk['Sales'] = df_uk['Quantity'] * df_uk['UnitPrice']
df_uk.head()
```

```python
df_user = pd.DataFrame(df['CustomerID'].unique(), columns=['CustomerID'])
df_user.head(3)
```

```python
tmp = df_uk.groupby(df_uk['InvoiceDate'].dt.date)['CustomerID'].count().reset_index()
tmp.columns = ['InvoiceDate', 'Count']
tmp.head()
```

```python
tmp = df_uk.groupby(df_uk['InvoiceDate'].dt.date)['Sales'].sum().reset_index()
tmp.columns = ['InvoiceDate', 'Sales']
tmp.head()
```

```python
last_purchase = df_uk.groupby('CustomerID')['InvoiceDate'].max().reset_index()
last_purchase.columns = ['CustomerID', 'LastPurchaseDate']
last_purchase.head()
```

```python
last_purchase['Retention'] = (last_purchase['LastPurchaseDate'].max() - last_purchase['LastPurchaseDate']).dt.days
last_purchase.head()
```

```python
df_user = pd.merge(df_user, last_purchase, on='CustomerID')
df_user = df_user[['CustomerID', 'Retention']]
df_user.head()
```

```python
df_retention = df_user[['Retention']]
df_retention.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_user['RetentionScaled'] = scaler.fit_transform(df_retention)
df_user.head()
```

```python
df_retention_scaled = df_user[['RetentionScaled']]
df_retention_scaled.head()
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, max_iter=1000)
    kmeans.fit(df_retention_scaled)
    wcss.append(kmeans.inertia_)

wcss = pd.DataFrame(data=np.c_[range(1, 10), wcss], columns=['NumberOfClusters', 'WCSS'])
wcss
```

```python
df_user['Cluster'] = kmeans.labels_
df_user.head()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.03, min_samples=5)
dbscan.fit(df_retention_scaled)
clusters = dbscan.labels_
df_user['Cluster'] = clusters
df_user.head()
```
